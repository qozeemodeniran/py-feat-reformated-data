```python
import pandas as pd
import os

# Define the directory path containing the CSV files
directory_path = "/Users/qozeemodeniran/Library/CloudStorage/GoogleDrive-qo00109@georgiasouthern.edu/.shortcut-targets-by-id/1kRnniww0E8Wk3_2jaztlV5tU1dWK_VpQ/Qozeem/py-feat reformatted data/Reformatted_PyFeat_Data/F1_Lorna"  # Replace with the directory containing your CSV files

# Define the prefix of the output CSV file name
output_file_prefix = "output_data_"

# List all CSV files in the directory
csv_files = [f for f in os.listdir(directory_path) if f.endswith(".csv")]

# Initialize an empty list of dictionaries for the final result
final_data = []

# Extract and append data from each CSV file
for idx, csv_file in enumerate(csv_files, start=2):
    # Read the current CSV file into a DataFrame
    data_df = pd.read_csv(os.path.join(directory_path, csv_file))
    
    # Extract the 'anger' column data
    anger_data = data_df['anger'].values.tolist()
    
    # Create a new row dictionary for the current participant (P2, P3, etc.)
    row_data = {"participant": f"P{idx}", **{f"T{i}": value for i, value in enumerate(anger_data, start=1)}}
    
    # Append the new row data to the final list
    final_data.append(row_data)

# Create a DataFrame from the list of dictionaries
final_df = pd.DataFrame(final_data)

# Define the output CSV file path
output_csv_path = "final_output_data.csv"

# Save the final DataFrame to a new CSV file
final_df.to_csv(output_csv_path, index=False)

print(f"CSV file '{output_csv_path}' created successfully!")
```
